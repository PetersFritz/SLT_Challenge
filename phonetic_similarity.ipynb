```python
import os
import glob
from Levenshtein import distance as levenshtein_distance
```

```python
def get_phonetic_similarity_edges(phonetic_transcript):
    words = phonetic_transcript.split()

    if len(words) < 2:
        return {}

    edges = {}

    for i in range(len(words) - 1):
        word1 = words[i]
        word2 = words[i + 1]

        sim = 1 - (levenshtein_distance(word1, word2) / max(len(word1), len(word2)))
        edges[(word1, word2)] = sim

    return edges

def process_phonetic_files(root_folder):
    pattern = os.path.join(root_folder, '**', '*.txt')
    file_paths = glob.glob(pattern, recursive=True)

    print(file_paths)
    
    results = {}
    for file_path in file_paths:
         with open(file_path, 'r') as file:
            phonetic_transcript = file.read().strip()
            edges = get_phonetic_similarity_edges(phonetic_transcript)

            results[file_path] = edges
            
    return results
```

```python
if __name__ == "__main__":
    base_path = "C:/Users/Student/Documents/phd_work/SLT_challenge"
    input_folder = base_path + "/" + "phonetic_outputFiles/PFT_outputFiles"

    results = process_phonetic_files(input_folder)

    print(results)

    for file_path, edges in results.items():
        print(f"File: {file_path}")
        print("Edges:")
        for pair, similarity in edges.items():
            print(f"  {pair}: {similarity}")
            print()
```
